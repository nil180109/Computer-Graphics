# DDA LINE DROWING

```python
import numpy as np
import matplotlib.pyplot as plt
def DDALine(x1, y1, x2, y2):
  print(x1)
  dx = x2 - x1
  dy = y2 - y1

  slop=dy/dx;
  x = np.array([x1])
  y = np.array([y1])


  while x1!=x2:
      if abs(dx)>=abs(dy):
          x1=x1+1;
          y1=y1+slop;
      else:
          x1=x1+1/slop
          y1=y1+1
      x = np.append(x, round(x1))
      y = np.append(y, round(y1))

  print(x)
  print(y)
  plt.scatter(x,y,color='g')
  plt.plot(x,y,color='r')
  plt.show()
  



# x1 = int(input("Enter X1: "))
# y1 = int(input("Enter Y1: "))
# x2 = int(input("Enter X2: "))
# y2 = int(input("Enter Y2: "))

x1=1
y1=2
x2=6
y2=7
DDALine(x1,y1,x2,y2)
```

### Bresenham

```python
import numpy as np
import matplotlib.pyplot as plt

def BresenhamLine(x1, y1, x2, y2):
  print(x1)
  dx = x2 - x1
  dy = y2 - y1
  p=2*dy-dx
  x = np.array([x1])
  y = np.array([y1])


  while x1!=x2:
      if p>=0:
        x1=x1+1
        y1=y1+1
        p=p+2*dy-2*dx

      else:
        x1=x1+1
        y1=y1
        p=p+2*dy

      x = np.append(x, x1)
      y = np.append(y, y1)

  print(x)
  print(y)
  plt.scatter(x,y,color='g')
  plt.plot(x,y,color='r')
  plt.show()


# x1 = int(input("Enter X1: "))
# y1 = int(input("Enter Y1: "))
# x2 = int(input("Enter X2: "))
# y2 = int(input("Enter Y2: "))

x1=9
y1=12
x2=14
y2=20
BresenhamLine(x1,y1,x2,y2)
```

### Midpoint Line Drawing Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

def MidPointLine(x1, y1, x2, y2):
  print(x1)
  dx = x2 - x1
  dy = y2 - y1
  d=2*dy-dx
  d0=2*dy-2*dx
  x = np.array([x1])
  y = np.array([y1])


  while x1!=x2:
      if d>=0:
        x1=x1+1
        y1=y1+1
        d=d+d0

      else:
        x1=x1+1
        y1=y1
        d=d+2*dy

      x = np.append(x, x1)
      y = np.append(y, y1)

  print(x)
  print(y)
  plt.scatter(x,y,color='g')
  plt.plot(x,y,color='r')
  plt.show()


# x1 = int(input("Enter X1: "))
# y1 = int(input("Enter Y1: "))
# x2 = int(input("Enter X2: "))
# y2 = int(input("Enter Y2: "))

x1=9
y1=14
x2=16
y2=20
MidPointLine(x1,y1,x2,y2)
```

### Midpoint Circle Drawing Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

def MidPointCircle(cx, cy, r):
  # Step-1
  x0=0
  y0=r
  #  Step-2
  p0=1-r

  # Step-3
  x = np.array([])
  y = np.array([])
  while y0>x0:
    x = np.append(x, x0)
    y = np.append(y, y0)
    if p0<0:
      x0=x0+1;
      y0=y0
      p0=p0+2*x0+1
    else:
      x0=x0+1
      y0=y0-1
      p0=p0-2*y0+2*x0+1


  l=x.size
  while l!=0:
    l=l-1
    x = np.append(x, y[l])
    y = np.append(y, x[l])

  l=x.size

# (x,-y)
  for i in range(l-1,-1,-1):
    x = np.append(x, x[i])
    y = np.append(y, -y[i])
# (-x,-y)
  for i in range(0,l,1):
    x = np.append(x, -x[i])
    y = np.append(y, -y[i])
  
# (-x,y)
  for i in range(l-1,-1,-1):
    x = np.append(x, -x[i])
    y = np.append(y, y[i])

  x = np.append(x, 0)
  y = np.append(y, r)
  l=x.size
  for i in range(0,l,1):
    x[i]=x[i]+cx;
    y[i]=y[i]+cy
  
  l=x.size
  print(l)
  print(x)
  print(y)

  plt.scatter(x,y,color='g')
  plt.plot(x,y,color='r')
  plt.show()


# x1 = int(input("Enter X1: "))
# y1 = int(input("Enter Y1: "))
# x2 = int(input("Enter X2: "))
# y2 = int(input("Enter Y2: "))

cx=0
cy=8
r=10
MidPointCircle(cx,cy,r)
```

### Boundary Fill Algorithm

```python
import numpy as np
import cv2


def boundary_fill(image, x, y, fill_color, boundary_color):
    # print(fill_color)
    # print(np.all(image[x, y] == fill_color))
    if not (np.all(image[x, y] == fill_color) or np.all(image[x, y] == boundary_color)):
        # print("shamim")
        image[x, y] = fill_color
        
        boundary_fill(image, x - 1, y, fill_color, boundary_color)
        boundary_fill(image, x, y + 1, fill_color, boundary_color)
        boundary_fill(image, x, y - 1, fill_color, boundary_color)
        boundary_fill(image, x + 1, y, fill_color, boundary_color)
        
xsize=50
ysize=50
image = np.zeros((xsize, ysize, 3), dtype=np.uint8)
image[0:50, 0:5] = (255, 0, 0)  # Red rectangle
image[0:5, 0:50] = (255, 0, 0)  # Red rectangle
image[0:50, 45:50] = (255, 0, 0)  # Red rectangle
image[45:50, 0:50] = (255, 0, 0)  # Red rectangle

boundary_fill(image, int(xsize/2), int(ysize/2), (0, 255, 0), (255, 0, 0))
plt.imshow(image)
plt.show()   
# print(image)
```

### Scan line fill Algorithm

```python
import turtle

def draw_polygon(vertices):
    turtle.penup()
    turtle.goto(vertices[0][0], vertices[0][1])
    turtle.pendown()

    for x, y in vertices[1:]:
        turtle.goto(x, y)

    turtle.goto(vertices[0][0], vertices[0][1])

def scan_line_fill_polygon(vertices, fill_color):
    min_y = min(vertices, key=lambda p: p[1])[1]
    max_y = max(vertices, key=lambda p: p[1])[1]

    for y in range(min_y, max_y + 1):
        intersections = []

        for i in range(len(vertices)):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % len(vertices)]

            if (y1 <= y and y2 > y) or (y2 <= y and y1 > y):
                x_intersection = int(x1 + (y - y1) / (y2 - y1) * (x2 - x1))
                intersections.append(x_intersection)

        intersections.sort()
        for i in range(0, len(intersections), 2):
            x_start = intersections[i]
            x_end = intersections[i + 1]

            for x in range(x_start, x_end+5,5):
                turtle.penup()
                turtle.goto(x, y)
                turtle.pendown()
                turtle.color(fill_color)
                turtle.dot()

# Example usage:
turtle.speed("slow")  # Set the turtle drawing speed

# Define the vertices of a simple polygon
polygon_vertices = [(0, 0), (100, 0), (100, 100),(0, 100)]

# Draw the polygon
draw_polygon(polygon_vertices)

# Call the scan_line_fill_polygon function
scan_line_fill_polygon(polygon_vertices, "blue")

# Hide the turtle and display the window
turtle.hideturtle()
turtle.done()
```

### Line Clipping

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np


def find_position(x,y):
    
    if (x>=Wmin_valueX) and (x<=Wmax_valueX) and (y>=Wmin_valueY) and (y<=Wmax_valueY):
        return 0
    elif  (x>Wmax_valueX) and (y>Wmin_valueY) and (y<Wmax_valueY):
        return 2 
    elif (x>Wmin_valueX) and (x<Wmax_valueX) and (y<Wmin_valueY):
        return 4
    elif (x<Wmin_valueX)  and (y>Wmin_valueY) and (y<Wmax_valueY):
        return 1
    elif (x>Wmin_valueX) and (x<Wmax_valueX)  and (y>Wmax_valueY):
        return 8
    elif  (x>Wmax_valueX) and (y<Wmin_valueY) :
        return 6
    elif (x<Wmin_valueX)  and (y<Wmin_valueY) :
        return 5
    elif (x<Wmin_valueX)   and (y>Wmax_valueY):
        return 9
    elif  (x>Wmax_valueX)  and (y>Wmax_valueY):
        return 10

def show(xsize,ysize,wx,wy,points_x,points_y,f):
    
    array_2d = np.zeros((xsize, ysize, 3), dtype=np.uint8)
    fig, ax = plt.subplots()
    
    ax.imshow(array_2d, cmap='viridis', interpolation='nearest', extent=[0, array_2d.shape[1], 0, array_2d.shape[0]])

    rectangle = patches.Rectangle((wx, wy), xsize/2, ysize/2, linewidth=1, edgecolor='red', facecolor='none')
    ax.add_patch(rectangle)
    
    # Draw line
    ax.plot(points_x, points_y, color='blue', marker='o', label='Line')
    
    # Set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    if f==0:
        ax.set_title('Befor Clipping')
    else:
        ax.set_title('After Clipping')
        
    
    # Add legend
    ax.legend()
    # Show the plot
    plt.show()
    
def clipping(points_x,points_y,pos,f,Wmin_valueX,Wmax_valueX,Wmin_valueY,Wmax_valueY,xsize,ysize):
    slop= (points_y[1]-points_y[0])/(points_x[1]-points_x[0])
    if pos==1:
        if f==0:
            x1=Wmin_valueX
            y1=points_y[1]-slop*(points_x[1]-x1)
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmin_valueX
            y1=points_y[0]+slop*(x1-points_x[0])
            points_x[1]=x1
            points_y[1]=y1
            
            
    elif pos==2:
        print(f)
        if f==0:
            x1=Wmax_valueX
            y1=points_y[1]-slop*(points_x[1]-x1)
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmax_valueX
            y1=points_y[0]+slop*(x1-points_x[0])
            points_x[1]=x1
            points_y[1]=y1
    elif pos==4:
        print(f)
        if f==0:
            y1=Wmin_valueY
            x1=points_x[1]-((points_y[1]-y1)/slop)
            points_x[0]=x1
            points_y[0]=y1
        else:
            y1=Wmin_valueY
            x1=points_x[0]+((points_y[1]-y1)/slop)
            points_x[1]=x1
            points_y[1]=y1
    elif pos==8:
        print(f)
        if f==0:
            y1=Wmax_valueY
            x1=points_x[1]-((points_y[1]-y1)/slop)
            points_x[0]=x1
            points_y[0]=y1
        else:
            y1=Wmax_valueY
            x1=points_x[0]+((points_y[1]-y1)/slop)
            points_x[1]=x1
            points_y[1]=y1
    elif pos==9:
        print(f)
        if f==0:
            x1=Wmin_valueX
            y1=Wmax_valueY
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmin_valueX
            y1=Wmax_valueY
            points_x[1]=x1
            points_y[1]=y1
    elif pos==10:
        print(f)
        if f==0:
            x1=Wmax_valueX
            y1=Wmax_valueY
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmax_valueX
            y1=Wmax_valueY
            points_x[1]=x1
            points_y[1]=y1
    elif pos==5:
        print(f)
        if f==0:
            x1=Wmin_valueX
            y1=Wmin_valueY
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmin_valueX
            y1=Wmin_valueY
            points_x[1]=x1
            points_y[1]=y1
    elif pos==6:
        print(f)
        if f==0:
            x1=Wmax_valueX
            y1=Wmin_valueY
            points_x[0]=x1
            points_y[0]=y1
        else:
            x1=Wmax_valueX
            y1=Wmin_valueY
            points_x[1]=x1
            points_y[1]=y1
        
            
        
    

# Create a sample 2D array
windowx=[12,37,37,12]
windowy=[12,12,37,37]
points_x = [9, 40]
points_y = [40, 9]

Wmax_valueX = np.max(windowx)
Wmin_valueX = np.min(windowx)

Wmax_valueY = np.max(windowy)
Wmin_valueY = np.min(windowy)

xsize=2*(Wmax_valueX-Wmin_valueX)
ysize=2*(Wmax_valueY-Wmin_valueY)

show(xsize,ysize,windowx[0],windowy[0],points_x,points_y,0)




# After Clipping


new_pointx=[]
new_pointy=[]


#  Calculate Position......
a=find_position(points_x[0],points_y[0])
b=find_position(points_x[1],points_y[1])
print(a,b)

bit_or=a | b
if bit_or==0:
    print("completely inside..")
    show(xsize,ysize,windowx[0],windowy[0],points_x,points_y,1)
else:
    bit_and= a & b
    if bit_and !=0:
        print("completly outside...")
        show(xsize,ysize,windowx[0],windowy[0],new_pointx,new_pointx,1)
        
    else:
        print("clipping...")
        if a!=0:
            clipping(points_x,points_y,a,0,Wmin_valueX,Wmax_valueX,Wmin_valueY,Wmax_valueY,xsize,ysize)
        if b!=0:
            clipping(points_x,points_y,b,1,Wmin_valueX,Wmax_valueX,Wmin_valueY,Wmax_valueY,xsize,ysize)
        show(xsize,ysize,windowx[0],windowy[0],points_x,points_y,1)
            
    




```
